```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv(R"C:\Users\smrut\Downloads\Mall_Customers.csv")

# Select features for clustering
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]

# Step 1: Use Elbow Method to find optimal K
inertia = []
k_range = range(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertia.append(km.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (within-cluster sum of squares)")
plt.title("Elbow Method For Optimal k")
plt.grid(True)
plt.show()

# Step 2: Apply KMeans with optimal k (e.g., 5)
k_optimal = 5
model = KMeans(n_clusters=k_optimal, random_state=42)
df['Cluster'] = model.fit_predict(X)

# Step 3: Visualize clusters
plt.figure(figsize=(8, 5))
colors = ['red', 'blue', 'green', 'purple', 'orange']
for i in range(k_optimal):
    cluster = df[df['Cluster'] == i]
    plt.scatter(cluster["Annual Income (k$)"], cluster["Spending Score (1-100)"],
                label=f"Cluster {i}", color=colors[i])
    
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Customer Segments using KMeans")
plt.legend()
plt.grid(True)
plt.show()
```
